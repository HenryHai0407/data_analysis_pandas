```python
# Create a-read-csv loop and append the data into a defined dataframe
da = pd.DataFrame(columns=['Order ID','Product','Quantity Ordered','Price Each','Order Date','Purchase Address'])
for i in range(1,13):
    file1 = pd.read_csv(f'dataset/sales2019_{i}.csv')
    da = pd.concat([da, file1], ignore_index=True)
print(da)
```

```python
# Rename with underline
df = da.rename(columns={'Order ID': 'order_id',
                        'Product': 'product',
                        'Quantity Ordered': 'quantity_ordered',
                        'Price Each': 'price_each',
                        'Order Date': 'order_date',
                        'Purchase Address': 'purchase_address'
                        })
```

```python
# Extract 'city' column from 'address' column
purchase_address = '944 Walnut St, Boston, MA 02215'
purchase_address.split(',')[1].strip()
```

```python
# Function
def get_city_from_address(address):
    return address.split(',')[1].strip()

get_city_from_address(purchase_address)
```

```python
df['city'] = df['purchase_address'].apply(get_city_from_address)
```

```python
# Using lambda
df['city1'] = df['purchase_address'].apply(lambda x: x.split(',')[1].strip())
df
```

```python
# drop city1
df = df.drop(['city1'],axis=1)
```

```python
df
```

```python
city_sales = df.groupby('city')['total'].sum()
city_sales
```

```python
# Plot bar chart
city_sales.plot(kind='bar', figsize=(12,6), color='skyblue')
plt.title('Total Sales by City')
plt.xlabel('City')
plt.ylabel('Total Sales')
# plt.xticks(rotation=45)
# plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# What time should we display ads to maximize the likelihood of customer’s buying products?
df['hour'] = df['order_date'].dt.hour
df
```

```python
hour_sales = df.groupby('hour')['total'].sum()
hour_sales
```

```python
ax =hour_sales.plot(kind='line',marker='o',figsize=(10,5))

# Set x-ticks
ax.set_xticks(hour_sales.index)
ax.set_xticklabels(hour_sales.index, rotation=0)

# Add labels and grid
plt.title('Hourly Sales')
plt.xlabel('Hour of Day')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
product_sales = df.groupby('product')['total'].sum()
product_sales
```

```python
# Pie chart
top_products = product_sales.sort_values(ascending=False).head(10)

# Plot pie chart
plt.figure(figsize=(8,8))
top_products.plot(kind='pie', autopct='%1.1f%%', startangle=140)

plt.title('Top 10 Product Sales Share')
plt.ylabel('')  # Hide y-label
plt.tight_layout()
plt.show()
```

```python
def join_product(products):
    return ",".join(products)

join_product(['USB-C Charging Cable','Wired Headphones'])
```

```python
# Link products into one
duplicates['all_products'] = duplicates.groupby(duplicates.index)['product'].apply(join_product)
duplicates
```

```python
cleaned_duplicates = duplicates[['all_products']].drop_duplicates()
cleaned_duplicates
```
